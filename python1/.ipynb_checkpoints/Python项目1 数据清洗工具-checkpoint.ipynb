```python
import numpy as np
import pandas as pd
```

```python
property_data = pd.read_csv('./property-data.csv')
```

```python
property_data
```

```python
property_data.info()
```

```python
#缺失值处理
new_df = property_data.dropna()
```

```python
new_df
```

```python
#缺失值处理字段
new_sub_df = property_data.dropna(subset=['ST_NUM'])
```

```python
new_sub_df
```

```python
#缺省值填充
new_fill_df = property_data.fillna('123456')
```

```python
new_fill_df
```

```python
#指定列填充
new_y_df = property_data['ST_NUM'].fillna('123456')
```

```python
new_y_df
```

```python
#均值,中位数,众数填充
df_mean = property_data['ST_NUM'].mean()
new_mean_df = property_data[['ST_NUM']].fillna(df_mean)
```

```python
new_mean_df
```

```python
df_mode = property_data['ST_NUM'].mode()
new_mode_df = property_data['ST_NUM'].fillna(df_mode)
```

```python
new_mode_df
```

```python
df_median = property_data['ST_NUM'].median()
new_median_df = property_data['ST_NUM'].fillna(df_median)
```

```python
new_median_df
```
